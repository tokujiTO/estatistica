```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

a=0.05
nv = [45, 50, 55, 60, 65, 52, 58, 61, 70, 75]
nivp = [65.0, 66.5, 67.8, 69.2, 70.3, 68.0, 68.5, 69.0, 70.1, 71.0]

plt.scatter(nv, nivp, color='black')
plt.xlabel('Numero de veículos')
plt.ylabel('Nível de poluição')

regressao = LinearRegression()

regressao.fit(np.array(nv).reshape(-1, 1), np.array(nivp).reshape(-1, 1))
inclinacao = regressao.coef_[0][0]
intercepto = regressao.intercept_[0]

x = np.linspace(min(nv), max(nv), 50).reshape(-1, 1)
y = regressao.predict(x)
plt.plot(x, y, color='red')
plt.scatter(np.array(nv).mean(), np.array(nivp).mean(), color='blue')
plt.grid()


```
